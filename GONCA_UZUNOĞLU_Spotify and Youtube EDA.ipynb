# SPOTİFY VE YOUTUBEDEKİ EN ÇOK DİNLENEN 5 ALBÜM

```python
# Spotify'daki en çok dinlenen ilk 5 albümü bulma
spotify_top_albums = df.groupby('Album')['Stream'].sum().nlargest(5)

# YouTube'da en çok izlenen ilk 5 albümü bulma
youtube_top_albums = df.groupby('Album')['Views'].sum().nlargest(5)

# Her iki platformdaki toplam akışları birleştirme
combined_streams = spotify_top_albums.add(youtube_top_albums, fill_value=0)

# Toplam akışa göre ilk 5 albümü sıralama
top_5_albums = combined_streams.nlargest(5)

print("Spotify ve YouTube'da en çok dinlenen ilk 5 albüm:")
print(top_5_albums)
```

```python
import matplotlib.pyplot as plt

# Spotify'daki en çok dinlenen ilk 5 albümü bulma
spotify_top_albums = df.groupby('Album')['Stream'].sum().nlargest(5)

# YouTube'da en çok izlenen ilk 5 albümü bulma
youtube_top_albums = df.groupby('Album')['Views'].sum().nlargest(5)

# Her iki platformdaki toplam akışları birleştirme
combined_streams = spotify_top_albums.add(youtube_top_albums, fill_value=0)

# Toplam akışa göre ilk 5 albümü sıralama
top_5_albums = combined_streams.nlargest(5)

# Çubuk grafik oluşturma
plt.figure(figsize=(10, 6))
top_5_albums.plot(kind='bar', color='skyblue')
plt.title('Spotify ve YouTube\'da En Çok Dinlenen İlk 5 Albüm')
plt.xlabel('Albüm')
plt.ylabel('Toplam Akış')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# ALBÜM TÜRÜ DAĞILIMI

```python
import matplotlib.pyplot as plt

# Albüm türlerinin dağılımını hesaplayın
album_type_distribution = df['Album_type'].value_counts()

# Çubuk grafik oluşturma
plt.figure(figsize=(10, 5))
album_type_distribution.plot(kind='bar', color='RED')
plt.title('Albüm Türü Dağılımı', fontsize=15)
plt.xlabel('Albüm Türü', fontsize=12)
plt.ylabel('Sayı', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='x')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import NullFormatter

def formatter(x, pos):
    return str(round(x / 1e9, 1))

def millformatter(x, pos):
    return str(round(x / 1e6, 1)) 

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(20, 10))

sns.countplot(data=df, x='Album_type', ax=ax1)
ax1.set_title("Her albüm türü için toplam şarkı sayısı")
ax1.set_ylabel("Toplam şarkı sayısı")
ax1.set_xlabel("Albüm türleri")

most_view_album = df.groupby('Album_type').sum().sort_values('Views', ascending=False).reset_index()
sns.barplot(data=most_view_album, y='Views', x='Album_type', ax=ax2)

ax2.set_title("Albüm türlerine göre toplam görüntülenme sayısı")
ax2.set_ylabel("Toplam Görüntülenme Sayısı (Milyar)")
ax2.set_xlabel("Albüm türleri")
ax2.yaxis.set_major_formatter(formatter)
ax2.yaxis.set_minor_formatter(NullFormatter())

fig.suptitle("Toplam albüm türleri")
plt.show()
```

# Youtube'da Dans Edilebilirliği yüksek en popüler video

```python
# Dans edilebilirliği yüksek olan en popüler videoları bulma
danceability_track = df.groupby(['Track', 'Danceability']).sum().reset_index()
danceability_track = danceability_track[danceability_track['Danceability'] > 0.9]\
    .sort_values('Views', ascending=False).head(10)

# Dans edilebilirliği yüksek olan en popüler videoları çubuk grafiği ile görselleştirme
fig = sns.barplot(data=danceability_track, y='Track', x='Views')
fig.set_title("Dans Edilebilirlik Oranı Yüksek Olan En Popüler Videolar (YouTube)", fontsize=15)
fig.xaxis.set_major_formatter(formatter)
fig.set_xlabel("Toplam Görüntüleme Sayısı (Milyarda)")
plt.show()
```

# Beğeni ve Yorumlar Bakımından YouTube  En İyi 10 Sanatçı

```python
# Sanatçılara göre beğeni ve yorum sayılarını toplamak için gruplama yapma
df_grouped = df.groupby('Artist')[['Likes', 'Comments']].sum().reset_index()
```

```python
# En çok beğeni alan ilk 10 sanatçıyı seçmek için sıralama yapma
df_grouped = df_grouped.sort_values(by='Likes', ascending=False)[:10]
```

```python
# Veriyi eritme (melt) işlemi uygulama
df_melted = pd.melt(df_grouped, id_vars=['Artist'], var_name='Platform', value_name='Counts')
```

```python
sns.barplot(y='Artist', x='Counts', hue='Platform', data=df_melted, orient = 'h')
plt.xticks(rotation=0)
plt.xlabel('Artist')
plt.ylabel('Counts')
plt.title('Top 10 Artists on YouTube and Spotify by Likes & Comments');
```

```python
# Görselleştirme
sns.barplot(y='Artist', x='Counts', hue='Platform', data=df_melted, orient='h', palette='husl')
plt.xticks(rotation=0)
plt.xlabel('Sanatçı')
plt.ylabel('Sayı')
plt.title('Beğeni ve Yorumlar Bakımından YouTube ve Spotify\'daki En İyi 10 Sanatçı');
```
