# SPOTƒ∞FY VE YOUTUBEDEKƒ∞ EN √áOK Dƒ∞NLENEN 5 ALB√úM

```python
# Spotify'daki en √ßok dinlenen ilk 5 alb√ºm√º bulma
spotify_top_albums = df.groupby('Album')['Stream'].sum().nlargest(5)

# YouTube'da en √ßok izlenen ilk 5 alb√ºm√º bulma
youtube_top_albums = df.groupby('Album')['Views'].sum().nlargest(5)

# Her iki platformdaki toplam akƒ±≈ülarƒ± birle≈ütirme
combined_streams = spotify_top_albums.add(youtube_top_albums, fill_value=0)

# Toplam akƒ±≈üa g√∂re ilk 5 alb√ºm√º sƒ±ralama
top_5_albums = combined_streams.nlargest(5)

print("Spotify ve YouTube'da en √ßok dinlenen ilk 5 alb√ºm:")
print(top_5_albums)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import NullFormatter

def formatter(x, pos):
    return str(round(x / 1e9, 1))

def millformatter(x, pos):
    return str(round(x / 1e6, 1)) 

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(20, 10))

sns.countplot(data=df, x='Album_type', ax=ax1)
ax1.set_title("Her alb√ºm t√ºr√º i√ßin toplam ≈üarkƒ± sayƒ±sƒ±")
ax1.set_ylabel("Toplam ≈üarkƒ± sayƒ±sƒ±")
ax1.set_xlabel("Alb√ºm t√ºrleri")

most_view_album = df.groupby('Album_type').sum().sort_values('Views', ascending=False).reset_index()
sns.barplot(data=most_view_album, y='Views', x='Album_type', ax=ax2)

ax2.set_title("Alb√ºm t√ºrlerine g√∂re toplam g√∂r√ºnt√ºlenme sayƒ±sƒ±")
ax2.set_ylabel("Toplam G√∂r√ºnt√ºlenme Sayƒ±sƒ± (Milyar)")
ax2.set_xlabel("Alb√ºm t√ºrleri")
ax2.yaxis.set_major_formatter(formatter)
ax2.yaxis.set_minor_formatter(NullFormatter())

fig.suptitle("Toplam alb√ºm t√ºrleri")
plt.show()
```

# Youtube'da Dans Edilebilirliƒüi y√ºksek en pop√ºler video

```python
# Dans edilebilirliƒüi y√ºksek olan en pop√ºler videolarƒ± bulma
danceability_track = df.groupby(['Track', 'Danceability']).sum().reset_index()
danceability_track = danceability_track[danceability_track['Danceability'] > 0.9]\
    .sort_values('Views', ascending=False).head(10)

# Dans edilebilirliƒüi y√ºksek olan en pop√ºler videolarƒ± √ßubuk grafiƒüi ile g√∂rselle≈ütirme
fig = sns.barplot(data=danceability_track, y='Track', x='Views')
fig.set_title("Dans Edilebilirlik Oranƒ± Y√ºksek Olan En Pop√ºler Videolar (YouTube)", fontsize=15)
fig.xaxis.set_major_formatter(formatter)
fig.set_xlabel("Toplam G√∂r√ºnt√ºleme Sayƒ±sƒ± (Milyarda)")
plt.show()
```

# Beƒüeni ve Yorumlar Bakƒ±mƒ±ndan YouTube  En ƒ∞yi 10 Sanat√ßƒ±

```python
# Sanat√ßƒ±lara g√∂re beƒüeni ve yorum sayƒ±larƒ±nƒ± toplamak i√ßin gruplama yapma
df_grouped = df.groupby('Artist')[['Likes', 'Comments']].sum().reset_index()
```

```python
# En √ßok beƒüeni alan ilk 10 sanat√ßƒ±yƒ± se√ßmek i√ßin sƒ±ralama yapma
df_grouped = df_grouped.sort_values(by='Likes', ascending=False)[:10]
```

```python
# Veriyi eritme (melt) i≈ülemi uygulama
df_melted = pd.melt(df_grouped, id_vars=['Artist'], var_name='Platform', value_name='Counts')
```

```python
sns.barplot(y='Artist', x='Counts', hue='Platform', data=df_melted, orient = 'h')
plt.xticks(rotation=0)
plt.xlabel('Artist')
plt.ylabel('Counts')
plt.title('Top 10 Artists on YouTube and Spotify by Likes & Comments');
```

```python
# G√∂rselle≈ütirme
sns.barplot(y='Artist', x='Counts', hue='Platform', data=df_melted, orient='h', palette='husl')
plt.xticks(rotation=0)
plt.xlabel('Sanat√ßƒ±')
plt.ylabel('Sayƒ±')
plt.title('Beƒüeni ve Yorumlar Bakƒ±mƒ±ndan YouTube ve Spotify\'daki En ƒ∞yi 10 Sanat√ßƒ±');
```
